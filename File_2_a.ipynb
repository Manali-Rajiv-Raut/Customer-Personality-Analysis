## 3. Check for missing Values

```python
marketing_data.isnull().sum()
```

#### Observation : We have only one column with null values i.e. income

```python
#finding the mean of the column to fill nan values
mean_value = marketing_data["Income"].mean()
mean_value
```

```python
median_value = marketing_data["Income"].median()
median_value
```

#### Observation:  Here I am not sure about the outliers if exist in the dataset. So filing null values with median 

```python
#filling mean values 
marketing_data["Income"].fillna(value = median_value , inplace = True)
```

```python
#checking again for null values
marketing_data.isnull().sum()
```

### Fantastic Job!!

## 4. Changing Dt_cutomer into DateTime Format

```python
marketing_data["Dt_Customer"]
```

```python
#convert the column to datetime format
marketing_data['Dt_Customer'] = pd.to_datetime(marketing_data['Dt_Customer'], dayfirst=True, infer_datetime_format = True)
```

```python
marketing_data["Dt_Customer"]
```

#### Obsrvation: dtype changed from object to datetimes64

## 5. Creating the new Columns

### 5.1 Age

```python
marketing_data["Age"] = 2022-marketing_data['Year_Birth']
```

```python
marketing_data["Age"].head(10)
```

### 5.2 Total Expenses

```python
marketing_data["Total_Expenses"] = marketing_data['MntWines'] + marketing_data['MntFruits'] + \
                                   marketing_data['MntMeatProducts'] + marketing_data['MntFishProducts'] + \
                                   marketing_data['MntSweetProducts'] + marketing_data['MntGoldProds']
```

```python
marketing_data["Total_Expenses"].head(10)
```

### 5.3 Total No. of Purchases

```python
#number of purchases the customer made
marketing_data["Total_Purchases"] = marketing_data["NumDealsPurchases"] + marketing_data["NumWebPurchases"] + \
                                    marketing_data["NumCatalogPurchases"] + marketing_data["NumStorePurchases"]
```

```python
marketing_data["Total_Purchases"].head(10)
```

### 5.3 Total Accepted Compaigns

```python
marketing_data["Total_accepted_campaign"] = marketing_data['AcceptedCmp1'] + marketing_data['AcceptedCmp2'] + \
                                            marketing_data['AcceptedCmp3'] + marketing_data['AcceptedCmp4'] +  \
                                            marketing_data['AcceptedCmp5'] + marketing_data['Response']
```

```python
marketing_data["Total_accepted_campaign"].head(10)
```

```python
marketing_data["Total_kids_home"] = marketing_data['Kidhome'] + marketing_data['Teenhome']
```

```python
marketing_data["Total_kids_home"].head(10)
```

## 6. Checking for Marital_status

```python
marketing_data["Marital_Status"].value_counts()
```

#### changing values 
- married = together
- single = alone = absurd = YOLO

```python
marketing_data['Marital_Status']= marketing_data.Marital_Status.replace({'Alone':'Single','Absurd':'Single','YOLO':'Single'})
marketing_data['Marital_Status'] = marketing_data.Marital_Status.replace({'Married':'Couple', 'Together':'Couple'})
```

```python
marketing_data["Marital_Status"].value_counts()
```

## 7. Checking for Duplicate values

```python
if marketing_data['ID'].nunique() == len(marketing_data):
    print("Unique Data")
else:
    print("Duplicates may Present")
```

### 9.10 Total_Expenses

```python
plt.figure(figsize=(10,10))
ax = marketing_data[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(). 
     \sort_values(ascending=True).plot(kind='barh')
plt.title('Total Expenses', pad=15, fontsize=18)
rects = ax.patches

for rect in rects:
    x_value = rect.get_width()
    y_value = rect.get_y() + rect.get_height() / 2
    plt.annotate('{}'.format(x_value), (x_value, y_value), xytext=(-49, 0),
    textcoords='offset points', va='center', ha='left', color = 'white', fontsize=11, fontweight='semibold')
```
